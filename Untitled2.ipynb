```python
import numpy as np
import math
import matplotlib.pyplot as plt
```

```python
def y_(x):
    return 60-2*math.exp(4-0.03*x)
```

```python
def dy_dx_(x):
    return 0.06*math.exp(4-0.03*x)
```

```python
x=np.linspace(0,150,1001)
x
```

```python
y=[]
for i in x:
    y.append(y_(i))
y=np.array(y)
y
```

```python
dy_dx=[]
for i in x:
    dy_dx.append(dy_dx_(i))
dy_dx=np.array(dy_dx)
dy_dx
```
